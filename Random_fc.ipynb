```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import nibabel as nib

# Generate synthetic time series data for 5 regions with 100 time points
np.random.seed(42)
n_regions = 5
n_timepoints = 100

# Create random time series data
time_series = np.random.rand(n_regions, n_timepoints)

# Introduce some correlations between regions
time_series[1] = time_series[0] + np.random.normal(0, 0.1, n_timepoints)  # ROI 1 and ROI 2 are correlated
time_series[3] = time_series[2] + np.random.normal(0, 0.1, n_timepoints)  # ROI 3 and ROI 4 are correlated

# Compute the correlation matrix
fc_matrix = np.corrcoef(time_series)

# Convert to DataFrame
fc_df = pd.DataFrame(fc_matrix, columns=[f'ROI_{i+1}' for i in range(n_regions)], index=[f'ROI_{i+1}' for i in range(n_regions)])

# Plot the functional connectivity matrix
plt.figure(figsize=(6, 5))
sns.heatmap(fc_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Functional Connectivity Matrix (Synthetic Data)')
plt.xlabel('Brain Regions')
plt.ylabel('Brain Regions')
plt.show()
```
